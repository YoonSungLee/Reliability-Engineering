* 난수를 만드는 방법
 * 컴퓨터를 사용하지 않는 방법: 동전 던지기, 주사위 던지기, 카드 섞기, 룰렛회전 등의 방법으로 진짜 난수를 얻을 수 있지만
   일반적으로 사용하기에 불편함이 많음
 * 컴퓨터를 사용하는 방법 <br>
1) Mid-Square method <br>
    예를 들어, 4자리 숫자를 만든다고 하자. 초기 숫자를 만든다. 예: 5232

    5232\*\*2 = 27,373,824, we get 3738, <br>
    3738\*\*2 = 13,972,644, we get 9726, <br>
    9726\*\*2 = 94,595,080, we get 5950, <br>

    과 같은 방법으로 4자리 숫자를 얻을 수 있다. 이 숫자는 진짜 난수는 아니지만, 난수처럼 사용할 수 있어 이를 의사난수, 가짜난수라고 부른다.Pseudo-random” or “Quasi-Random”  
    이후 이 숫자를 9999로 나누어서 0과 1 사이의 숫자를 출력한다.  
    하지만 숫자를 얻어내는 알고리즘이 노출되면 제 3자가 난수값을 알아낼 수 있다는 단점이 있다.<br>
    <br>
    2) Multiplicative generator<br>
   두번째 방법은 일반적으로 널리 사용되는 방법으로 multiplicative generator이다. <br>
   
   $ \ X_i = a X_{i-1} \% \thinspace m, \thinspace m = 2 ^\beta, \, a = 2^{17} +3, \beta = 35  $(32bit system 기준, 나머지가 없도록 상수 설정)
   
   $ X_0=YY+100*(MM-1+12*(DD-1+31*(HH+24* (Min+60 *SS)))) $  
   Y:년, M:월, D:일, H:시, Min:분, SS:밀리세컨드
   
   이 값의 최대값은 m 이므로 m으로 나누면 0~1 사이의 실수가 만들어 진다.  
   X ~ U(0,1) : X는 -과 1 사이의 임의의 실수(Uniform Distribution)

```python
import datetime
dt = datetime.datetime.now() # 현재시각
dt.year, dt.month, dt.day, dt.hour, dt.minute, dt.second, dt.microsecond

YY = dt.year
MM = dt.month
DD = dt.day
HH = dt.hour
Min = dt.minute
SS = dt.second

beta = 35
m = 2**beta
a = 2**17 + 3

seed = YY + 100 * (MM - 1 + 12 * (DD - 1 + 31*(HH + 24*(Min + 60 * SS))))
x = seed
for i in range(10):
    x = a*x % m
    print(x / m)
```

* Bernoulli 분포: p의 확률로 1을 가지는 확률변수는 Bernoulli(p)을 따른다. p는 성공확률이다. <br>
  $  X = \begin{cases} 1 & \text{when success} \\ 0 & \text{when fail} \end{cases} $  
  X ~ Bernoulli(p)
  

```python
import datetime
dt = datetime.datetime.now()
dt.year, dt.month, dt.day, dt.hour, dt.minute, dt.second, dt.microsecond

YY = dt.year
MM = dt.month
DD = dt.day
HH = dt.hour
Min = dt.minute
SS = dt.second

beta = 35
m = 2**beta
a = 2**17 + 3

seed = YY + 100 * (MM - 1 + 12 * (DD - 1 + 31*(HH + 24*(Min + 60 * SS))))
x = seed

p = 0.5
for i in range(10):
    x = a*x % m
    if x / m < p:
        rnd = 1
    else:
        rnd = 0
    print(rnd)    
```

이항분포를 따르는 난수 발생기를 클래스로 개발해 보자. 



```python
class binomial:
   
    def __init__(self, n, p):
        import datetime
        self.n = n
        self.p = p
        self.beta = 35
        self.m = 2**self.beta
        self.a = 2**17 + 3
        
    def get(self, size = 1):
        dt = datetime.datetime.now()
        YY = dt.year
        MM = dt.month
        DD = dt.day
        HH = dt.hour
        Min = dt.minute
        SS = dt.second
        
        seed = YY + 100 * (MM - 1 + 12 * (DD - 1 + 31*(HH + 24*(Min + 60 * SS))))
        x = seed
        
        res = []

        for i in range(size):
            rnd = 0
            for j in range(self.n):
                x = self.a*x % self.m
                if x / m < self.p:
                    rnd += 1
            res.append(rnd)
        return res
            
x = binomial(10,0.5)
result = x.get(100)
print(sum(result)/100)
```

```python
class binomial:
    def __init__(self,n,p):
        self.n = n
        self.p = p
        import datetime
        
    def get(self, size=1):
        dt = datetime.datetime.now()
        YY = dt.year
        MM = dt.month
        DD = dt.day
        HH = dt.hour
        Min = dt.minute
        SS = dt.second
        
        a = 2**17+3
        beta = 35
        m = 2**beta
        
        seed = YY+100*(MM-1+12*(DD-1+31*(HH+24*(Min+60*SS))))
        x = seed
        
        res = []
        
        for i in range(size):
            rnd = 0
            for j in range(self.n):
                x = a*x%m
                if x/m < self.p:
                    rnd +=1
            res.append(rnd)
        return res
```

### <font color='blue'> Prime number(소수) </font>  

* 일반적인 방법 <br>
  자기보다 작은 수로 나눠서 나눠지는 수가 존재하면 소수에서 제외한다.
  

```python
# general coding
primeList = [] # 소수를 보관할 리스트를 생성한다.

for i in range(2, 101): # 2~100 사이의 소수를 만들기 위해 ...
    primeList.append(i) # 해당 수가 소수라고 간주하고 ...
    for j in range(2,i):
        if i % j == 0:  # 해당 수가 소수가 아니면 소수 리스트에서 삭제한다.
            del primeList[-1]
            break

print(primeList)
```

* another algorithm <br>
  아래는 소수가 아니다.

    2 : 4, 6, 8, ... <br>
    3: 6, 9,12, ...<br>
    4: 8, 12, 16, ...<br>
   
   until 
    $\sqrt n $ 
    
    why $\sqrt n $ ?    두 수의 곱 a * b = n 이면, 반드시 min(a,b) <= sqrt(n)이 성립한다.그러므로, a값을 $\sqrt n $ 보다 작게 설정하는 것이다.
    

```python
is_prime = np.array(list(range(101)))
print(is_prime[4::2])
```

```python
# efficient coding(아리스토테네스의 채)
import numpy as np
is_prime = np.array(list(range(101)))
N_max = int(np.sqrt(len(is_prime) - 1)) # sqrt(n) 

for j in range(2, N_max + 1):
    is_prime[2*j::j] = 0 # is_prime[4::2], is_prime[6::3], is_prime[8::4], ...
```

![randomWalk](https://drive.google.com/uc?id=1gZkV0dv7rqs5C-B6QqQOcFFUoBzA3dMV)

랜덤워커가 N 스텝 이동 후, 평균적으로 어느 위치에 있을까? 이에 대한 수학적 해는  $\sqrt t $ 이다.
하지만, 이 문제의 수학적 해를 구하는 것은 어렵다.

```python
# 현재 위치에서 좌우로 이동하는 상황을 시뮬레이션 해보자.

np.random.randint(0,2,10) * 2 - 1 # [0 or 1] * 2 -1 = [-1,1]
```

```python
tMax = 200 # 200 스텝 움직임
nStories = 1000 # 기대값을 구하기 위해 1000회 반복
steps= np.random.randint(0, 2,(nStories, tMax)) * 2 - 1 
steps.shape
```

```python
print(steps[0:3])
```

```python
positions = np.cumsum(steps, axis=1) # axis = 1: 시간축으로 현재 위치를 알기 위해 누적합을 구함
```

```python
print(positions[0:3])
```

```python
sq_distance = positions**2
mean_sq_distance = np.mean(sq_distance, axis=0)
```

```python
distance = np.sqrt(mean_sq_distance)
print(distance) # Theoretically, sqrt (t).
```

```python
t = np.arange(tMax)
plt.plot(t, distance, 'g.', t, np.sqrt(t), 'y-') # simulation vs theoretical result
```

이론적인 해를 구해보자.





X: t번 이동해서 오른쪽으로 이동한 회수 $ X \sim Bin(t, 1/2) $ <br>
Y: t번 이동한 후 위치라고 하자.       $ Y = X - (t-X) = 2X - t$ <br>

$Y^2$ 은 t번 이동한 후 원점과의 거리의 제곱이 된다. 

$ E(Y^2) = E(4X^2 - 4tX + t^2) = 4E(X^2) -4tE(X) + t^2 $ <br>
$ V(X) = E(X^2)-E(X)^2 \rightarrow E(X^2) = V(X) + E(X)^2 = t/4 + t^2/4 $<br>

$ \therefore\  E(Y^2) = t + t^2 - 4t \frac {t}{2} + t^2 = t $

그러므로, 평균 위치는 $ \sqrt{t} $ 가 된다.
