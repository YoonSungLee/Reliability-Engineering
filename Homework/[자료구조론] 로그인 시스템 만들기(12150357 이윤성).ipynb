```python
import csv
import hashlib

def uName2int(uName):
    summ = ''
    for i in range(len(uName)):
        summ += str(ord(uName[i]))
    return int(summ)

def LoginSystem(id, pwd):
    _id = uName2int(id)
    if _id not in datalist:
        print("ID ERROR")
    else:
        _pwd = hashlib.sha256(pwd.encode()).hexdigest()
        if datalist[_id] != _pwd:
            print("PASSWORD ERROR")
        else:
            print("welcome " + id + "!")
            return True
        
datalist = {}
f = open("users.csv", "r", encoding = "utf-8")
rdr = csv.reader(f)
for line in rdr:
    datalist[uName2int(line[0])] = hashlib.sha256(line[1].encode()).hexdigest()
f.close()

# for lin in csv.reader(f):
#     users.append(line[0])
#     realPasswd.append(line[1])
#     _tmp = line[1].encode()
#     hashPasswd.append(hashlib.sha256(_tmp).hexdigets())
#     i += 1
```
