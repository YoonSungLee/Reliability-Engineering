```
import random
import time

class BNode:
    def __init__(self, item, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right
        
class BSTree:
    def __init__(self, root):
        self.root = root

    def insert(self, item):  # item을 삽입한다.
        self.root = self._insert(self.root, item)

    def _insert(self, curNode, item):
        # curNode 부터 밑으로 item 삽입위치를 찾아 삽입한다.
        # 현재 노드가 None이면 현재노드에 새노드를 삽입한다.
        if curNode == None:
            curNode = BNode(item)
        elif item < curNode.item:
        # 삽입할 값이 현재 노드값 보다 작으면 좌측을 현재노드로 보고 삽입을 실행한다.
            curNode.left = self._insert(curNode.left, item)
        else:
        # 삽입할 값이 현재 노드값 보다 크면 우측을 현재노드로 보고 삽입을 실행한다.
            curNode.right = self._insert(curNode.right, item)
        # 재귀가 끝나서 새노드를 삽입했다면 자신을 호출한 프로세스에 새노드를 리턴한다.
        return curNode

    def search(self, item):
        # 루트부터 item을 찾는다.
        return self._search(self.root, item)

    def _search(self, curNode, item):
        # curNode 밑에서 item을 찾는다.
        if curNode == None:
            print("item not found")
        elif curNode.item == item:
            # 찾은 item을 리턴한다.
            return curNode
        elif curNode.item > item:
            # 최종적으로 찾은 item을 리턴해야 한다. 리턴이 없으면 찾아도 리턴되지 않으므로 재귀를 끝낼때 리턴에 의해 item을 연쇄적으로 반환해야 한다.
            return self._search(curNode.left, item)
        elif curNode.item < item:
            # 최종적으로 찾은 item을 리턴해야 한다.
            return self._search(curNode.right, item)
```

```
numset = [x for x in range(1,100001)]
random.shuffle(numset) # 1부터 100000까지의 수를 섞어서 리스트로 정의
testnum = random.randint(1,100001) # 1부터 100000까지의 수 중 하나를 랜덤으로 선택
# testnum = numset[random.randint(1,100001)]
```

```
# 순차탐색 속도
ListTest = numset[:]
s = time.time()
for i in ListTest:
    if i == testnum:
        print(i)
        break
print("순차탐색시간 :", time.time() - s)
```

* 리스트는 탐색시간의 variance가 크다.
* 파이썬 언어 -> 수많은 라이브러리를 편리하게 사용 가능 but 컴퓨터의 시간 포기(Optimal X), 특정한 주제를 실험하기 위한 용도
* 상대적으로 리스트는 삽입/삭제가 빠르고, 이진탐색트리는 탐색이 빠르다. 그리고 데이터가 많을수록 그 차이는 크다.

```
# 이진탐색트리 속도
BSTreeTest = BSTree(BNode(numset[0]))
for i in numset[1:]:
    BSTreeTest.insert(i)
s = time.time()
print(BSTreeTest.search(testnum).item)
print("이진탐색트리탐색시간 :",time.time() - s)
```
