```
# Queue 클래스

class Queue:
    def __init__(self):
        self.q = []
        
    def enQueue(self, item):
        self.q.append(item)
        
    def deQueue(self):
        if self.isEmpty() == False: 
            return self.q.pop(0)
    
    def size(self):
        return len(self.q)
    
    def isEmpty(self):
        if len(self.q) > 0:
            return False
        else:
            return True
        
    def peek(self):
        if self.isEmpty() == False: 
            return self.q[0]
    
    def delete(self, item):
        if item in self.q: 
            self.q.remove(item)
        else:
            print("해당 아이템이 존재하지 않습니다.")
```

```
# Cust 클래스

class Cust:
    def __init__(self):
        self.arriveTime = 0
        self.orderTime = 0
        self.outTime = 0

# Shop 클래스
        
class Shop:
    def __init__(self):
        self.custQueue = Queue()
        
    def getSize(self):
        return self.custQueue.size()
    
    def entCust(self,cust):
        self.custQueue.enQueue(cust)
        
    def outCust(self):
        if self.getSize() != 0:
            while self.custQueue.peek().outTime < curTime:
                self.custQueue.deQueue()
                if self.getSize() == 0:
                    break
            
    def getLast(self):
        return self.custQueue.q[self.getSize()-1]
```

```
# 주문을 두명이 받는다면 어떻게 되는가?

import numpy as np
np.random.seed(seed=1)
randomTime = np.random.exponential(1)


shop1 = Shop()
shop2 = Shop()
id = 0
waiting_people_count1 = []
waiting_people_count2 = []


while randomTime < 840:
    print("고객 id : %d"%id, end = ' ')
#     print(id, end = '  ')
    person = Cust()
    person.arriveTime = randomTime
    print("도착시간 : {}".format(person.arriveTime), end = ' ')
#     print(person.arriveTime, end = '  ')
    curTime = person.arriveTime
    shop1.outCust()
    shop2.outCust()
    if shop1.getSize() <= shop2.getSize():
        shop = shop1
    else:
        shop = shop2
    if shop.getSize() == 0:
        person.orderTime = person.arriveTime
    else:
        person.orderTime = shop.getLast().outTime
    print("주문시간 : {}".format(person.orderTime), end = ' ')
#     print(person.orderTime, end = '  ')
    cookTime = np.random.normal(1, 0.2)
    while cookTime <0:
        cookTime = np.random.normal(1, 0.2)
    person.outTime = person.orderTime + cookTime
    print("나간시간 : {}".format(person.outTime), end = ' ')
#     print(person.outTime, end = '  ')
    shop.entCust(person)
    if shop == shop1:
        waiting_people_count1.append(shop.getSize())
    else:
        waiting_people_count2.append(shop.getSize())
    print("shop1 :",shop1.getSize(), end = '  ')
    print("shop2 :",shop2.getSize())
    id += 1
    randomTime += np.random.exponential(1)
```
