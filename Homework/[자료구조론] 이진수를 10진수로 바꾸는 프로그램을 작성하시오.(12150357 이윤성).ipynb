```
def change(x):
    x = x[::-1]
    answer = 0
    for i,number in enumerate(x):
        answer += int(number)*pow(2,i)
    return answer

x = input('2진수를 입력하세요 : ')
print('변환된 값은 %d입니다.'%change(x))
```
