```
# 클래스 이름은 대문자로 시작
class Node:
  def __init__(self, item = None): # 생성자(예약어)
    self.item = item
    self.link = None

a = Node('사과')
print(a.item, a.link)
```

```
class LinkedList: # LinkedList == Linked_list
  def __init__(self):
    self.root = Node()
  
  def append(self, item):
    newNode = Node(item)
    if self.root.item == None:
      self.root = newNode
    else:
      curNode = self.root
      while curNode.link != None:
        curNode = curNode.link # curNode.link는 다음 Node의 item 주소를 가리키고 있으므로 curNode는 다음 Node의 item이 된다
      curNode.link = newNode
      
  def remove(self, item):
    curNode = self.root
    preNode = self.root
    if self.root.item == item:
      self.root = self.root.link
    else:
      while curNode != None:
        if curNode.item == item:
          preNode.link = curNode.link
        preNode = curNode
        curNode = curNode.link
    
  def size(self):
    size = 1
    curNode = self.root
    while curNode.link != None:
      size += 1
      curNode = curNode.link
    return size
  
  def insert(self,idx, item):
    if idx < 0 or idx >= self.size(): #0번째부터 실행가능하다고 가정
      # 두 함수는 독립적인데, 하나의 함수에 다른 함수를 이용하기 위해 self.method() 이용
      print("인덱스 오류입니다")
    elif idx == self.size():
      self.append(item)
    else:
      newNode = Node(item)
      curidx = 0
      curNode = self.root
      for i in range(idx-1):
        curNode = curNode.link
      _tmp = curNode.link #_tmp : 의미없는 임시 변수
      curNode.link = newNode
      newNode.link = _tmp
  
  def print(self):
    curNode = self.root
    print(curNode.item, end = '-')
    while curNode.link != None:
      curNode = curNode.link
      print(curNode.item, end = '-')
    

```

```
fruits = LinkedList()
fruits.append('사과')
fruits.append('앵두')
fruits.append('배')
fruits.append('포도')
fruits.print()
print(fruits.size()) #'size()'의 괄호는 fruits의 메소드를 의미
fruits.insert(2, '귤')
fruits.print()
```
