```
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

```
def f(x):
    return x**2 + 10*np.sin(x)

x = np.arange(-10, 10, 0.1)
plt.plot(x, f(x))
plt.show() # 이 함수의 최소(Global Minimum)는 -1.3 근방이고, 지역 최소(Local Minimum)는 3.8 근방이다.
```

```
def BisectionMethod(f, a, b):
    interval = (b-a)/2
    m = a + interval
    while interval > 10e-6:
        if (f(a) > f(m)) and (f(b) > f(m)):
            b = m
            m = (a+b)/2
            interval = interval/2
        else:
            if f(a) > f(m) > f(b):
                a += interval
                m = a+interval
                b = m+interval
            else:
                a -= interval
                m = a+interval
                b = m+interval
    return m        
```

```
BisectionMethod(f,-2.5,1) # Global Minimum
```

```
BisectionMethod(f,2.5,6) # Local Minimum
```

```
# learning_rate, dx, abs(x_new-x_old) 설정에 대한 리스크가 존재한다.

def GradientDescentAlgorithm(f, x_old, dx=1e-6, learning_rate = 1e-5):
    x_new = x_old - learning_rate*derivative(f,x_old,dx)
    while abs(x_new-x_old) > 1e-8:
        x_old = x_new
        x_new = x_old - learning_rate*derivative(f,x_old,dx)
    return x_new
```

```
GradientDescentAlgorithm(f, 1) # Global Minimum
```

```
GradientDescentAlgorithm(f, 6) # Local Minimum
```

```
def f(x):
    return x**2 + 10*np.sin(x)

def f_prime(x):
    return 2*x + 10*np.cos(x)

def GradientDescentAlgorithm(f_prime, x_old, learning_rate = 1e-5):
    x_new = x_old - learning_rate*f_prime(x_old)
    while abs(x_new-x_old) > 1e-8:
        x_old = x_new
        x_new = x_old - learning_rate*f_prime(x_old)
    return x_new
```

```
GradientDescentAlgorithm(f_prime, 1) # Global Minimum
```

```
GradientDescentAlgorithm(f_prime, 6) # Local Minimum
```
