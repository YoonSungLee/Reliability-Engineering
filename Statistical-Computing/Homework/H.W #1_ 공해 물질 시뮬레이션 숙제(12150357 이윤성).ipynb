```
import numpy as np
```

```
class Dust:
    def __init__(self, name, x=76,y=-209):
        self.name = name
        self.x = x
        self.y = y
    def move(self, step):
        direction_list = np.random.rand(step)
        for rnd in direction_list:
            death_probability = np.random.rand()
            if death_probability < 1/4000:
                return 0
            else:
                if rnd<0.4:
                    self.x += 1
                elif rnd<0.5:
                    self.x -= 1
                elif rnd<0.7:
                    self.y += 1
                else:
                    self.y -= 1
                if (self.x > 600) and (-500 < self.y < 0):
                    return 1
        return 0
```

```
def Calculator(nStories, tMax):
    dust_list = [Dust(i) for i in range(nStories)]
    influence_count = 0
    for dust in dust_list:
        influence_count += dust.move(tMax)
    print('설정한 %d개의 입자 중에서 한국에 영향을 미친 입자의 개수는\
    %d개이며, 비율은 %f입니다.'%(nStories, influence_count, influence_count/nStories))
```

```
nStories = 10000
tMax = 10000 # 권장

Calculator(nStories, tMax)
```
