```
import numpy as np
import matplotlib.pyplot as plt
```

```
x_data = np.arange(1,13)
y_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
y_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
print('x_data ',x_data)
print('y_max', y_max)
print('y_min', y_min)
```

```
from scipy import optimize

def func(x, a, b, c):
    return a*(x**2) + b*x + c
```

```
# max temperature
params, params_covariance = optimize.curve_fit(func, x_data, y_max)
# print(params)
max_yHat = func(x_data, *params)
plt.scatter(x_data, y_max)
plt.plot(x_data, max_yHat)
```

```
# min temperature
params, params_covariance = optimize.curve_fit(func, x_data, y_min)
# print(params)
min_yHat = func(x_data, *params)
plt.scatter(x_data, y_min)
plt.plot(x_data, min_yHat)
```

```
# data list by curve_fit
print('max temperature\n',np.array(max_yHat).reshape(-1,1))
print('min temperature\n',np.array(min_yHat).reshape(-1,1))
```
